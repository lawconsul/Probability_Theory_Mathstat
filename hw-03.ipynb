```python
import numpy as np
import pandas as pd
```

```python
from math import factorial as fl
def C(k,n):
    return fl(n)/(fl(k)*fl(n-k))

def fB(k,n,p):
    return C(k,n) * p**k * (1-p)**(n-k)

n = 0
p = 0
lambda_ = n*p
def P(m, lambda_):
    P_m = lambda_** m / fl(m) * np.exp(-lambda_)
    return P_m
def Pnp(m, n, p):
    lambda_ = n*p
    P_m = lambda_** m / fl(m) * np.exp(-lambda_)
    return P_m
```

```python
# Даны значения зарплат из выборки выпускников: 
# 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. 
# Посчитать (желательно без использования статистических методов наподобие std, var, mean) 
# среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.
a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
s = pd.Series(a)
s.describe()
```

```python
mean = s.sum()/s.count()

def get_sko_arr(s,mean):
    m = np.empty((s.count()))
    for i in range(s.count()):
        m[i] = (s[i]-mean)**2
    return m

def get_sko(std_arr):
    n = sko_arr.size
    sko = 0
    for i in sko_arr:
        sko += i
    sko = (sko/(n-1))**0.5
    return sko

def get_nod(std_arr):
    n = sko_arr.size
    sko = 0
    for i in sko_arr:
        sko += i
    nod = sko/(n-1)
    return nod

def get_sod(std_arr):
    n = sko_arr.size
    sko = 0
    for i in sko_arr:
        sko += i
    sod = sko/n
    return sod

sko_arr = get_sko_arr(s,mean)
sko = get_sko(std_arr)
nod = get_nod(std_arr)
sod = get_sod(std_arr)

print(f'Среднее арифметическое: {mean}')
print(f'Среднее квадратичное отклонение: {sko}')
print(f'Несмещенная оценочная дисперсия: {nod}')
print(f'Cмещенная оценочная дисперсия: {sod}')
```

```python
# В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. 
# Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?
P_3whiteball = C(0,5)/C(2,8)*C(3,5)/C(4,12) + C(1,5)/C(2,8)*C(2,5)/C(4,12) + C(2,5)/C(2,8)*C(1,5)/C(4,12) 
print(f'Вероятность того, что 3 мяча белые: P = {P_3whiteball}')
```

```python
# На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. 
# Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. 
# Найти вероятность того, что выстрел произведен: 
P_A = 1/3*0.9 + 1/3*0.8 + 1/3*0.6
# a). первым спортсменом 
P_B1 = 1/3*0.9 / P_A
print(f'Вероятность того, что выстрел произведен первым спортсменом: P = {P_B1}')
# б). вторым спортсменом 
P_B2 = 1/3*0.8 / P_A
print(f'Вероятность того, что выстрел произведен вторым спортсменом: P = {P_B2}')
# в). третьим спортсменом.
P_B3 = 1/3*0.6 / P_A
print(f'Вероятность того, что выстрел произведен третьим спортсменом: P = {P_B3}')
```

```python
# В университет на факультеты A и B поступило равное количество студентов, 
# а на факультет C студентов поступило столько же, сколько на A и B вместе. 
# Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. 
# Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. 
P_A = 1/4*0.8 + 1/4*0.7 + 1/2*0.9
# Студент сдал первую сессию. Какова вероятность, что он учится: 
# a). на факультете A 
P_B1 = 1/4*0.8/P_A
print(f'Вероятность того, что он учится на факультете A: P = {P_B1}')
# б). на факультете B 
P_B2 = 1/4*0.7/P_A
print(f'Вероятность того, что он учится на факультете В: P = {P_B2}')
# в). на факультете C?
P_B3 = 1/2*0.9/P_A
print(f'Вероятность того, что он учится на факультете С: P = {P_B3}')
```

```python
# Устройство состоит из трех деталей. 
# Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. 
# Какова вероятность того, что в первый месяц выйдут из строя: 
# а). все детали 
P3 = 0.1*0.2*0.25
print(f'Вероятность того, что в первый месяц выйдут из строя все детали: P = {P3}')
# б). только две детали 
P2 = 0.1*0.2 + 0.2*0.25 + 0.1*0.25
print(f'Вероятность того, что в первый месяц выйдут из строя только две детали: P = {P2}')
# в). хотя бы одна деталь 
P1plus = 1 - (1-0.1)*(1-0.2)*(1-0.25)
print(f'Вероятность того, что в первый месяц выйдут из строя хотя бы одна деталь : P = {P1plus}')
# г). от одной до двух деталей?
P12 = P1plus - P3
print(f'Вероятность того, что в первый месяц выйдут из строя от одной до двух деталей: P = {P12}')
```
